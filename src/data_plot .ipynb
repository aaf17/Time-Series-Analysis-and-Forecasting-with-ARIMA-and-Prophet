```python
# Importing the libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the clean data

cleanData = pd.read_csv("Dataset/clean_data.csv")
cleanData.head()
```

```python
# Plotting the data

cleanData.plot(x='Date' , y='Price', figsize=(20,8), grid=True)
```

```python
# Detailed analysis of data

cleanData.describe()
```

```python
# Creating a column with month and year

cleanData["Year"] = pd.to_datetime(cleanData["Date"]).dt.year 
cleanData["Month"] = pd.to_datetime(cleanData["Date"]).dt.month
cleanData.head()
```

```python
# Plotting the heat map

plot = cleanData.pivot_table('Price', 'Year', 'Month')
heatMap = sns.heatmap(plot)
```
